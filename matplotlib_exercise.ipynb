## 1問目
使うcsvはhist_data.csvです。下にコードを書いてください。

![問1](./imgs/matplotlib/plot_1.png)

```python
hist_data = pd.read_csv('C:/Users/Cowlisa/PG/BasicPython/data/hist_data.csv ')
fig, ax = plt.subplots()

ax.hist(hist_data['aデータ'], bins = 30, color='orange', alpha=0.9, label='aデータ')
ax.hist(hist_data['bデータ'], bins = 30, color='skyblue', alpha=0.7, label='bデータ')
ax.set_title('aデータとbデータの温度に関する度数分布')
ax.set_xlabel('温度(℃)')
ax.set_ylabel('度数')
plt.legend()
ax.grid(True)
plt.show
```

## 2問目
使うcsvはscatter_data.csvです。下にコードを書いてください。

![問2](./imgs/matplotlib/plot_2.png)

```python
scatter_data = pd.read_csv('C:/Users/Cowlisa/PG/BasicPython/data/scatter_data.csv')
plt.scatter(scatter_data['length (cm)'],scatter_data['width (cm)'])
plt.grid(True)
plt.xlabel("length(cm)")
plt.ylabel("width(cm)")
plt.title("クラス0～2のlengthとwidthに関する散布図")

for i in range(len(scatter_data)):
    if scatter_data["class"][i] == 0:
        plt.scatter(scatter_data['length (cm)'][i],scatter_data['width (cm)'][i],color = "r", label = "class0" if i == 0 else "")
    elif scatter_data["class"][i] == 1:
        plt.scatter(scatter_data['length (cm)'][i],scatter_data['width (cm)'][i],color = "b", label = "class1" if scatter_data["class"][i - 1] != 1 else "")
    elif scatter_data["class"][i] == 2:
        plt.scatter(scatter_data['length (cm)'][i],scatter_data['width (cm)'][i],color = "g", label = "class2" if scatter_data["class"][i - 1] != 2 else "")

plt.legend()
plt.show()
```
